```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import scipy as sp
import pandas as pd
import random
```

```python
with open("./OutputPPP/out_Ian10.txt", "r") as fp:
    lineas = fp.read().split("\n")
```

```python
matrix = []
experiment = []

caract_matrix = []
caract = []

flag = True
for linea in lineas:
    if flag:
        caract.append(linea)
        flag = False
        continue
    try:
        experiment.append(int(linea));
    except:
        if "BEST" in linea or "Time:" in linea:
            caract.append(linea)
            continue
        caract_matrix.append(caract)
        matrix.append(experiment)
        experiment = []
        caract = []
        caract.append(linea)
```

```python
for i in range(len(caract_matrix)):
    print(str(i) + ":" + str(caract_matrix[i]))
```

```python
def resets_quality(matrix):
    best_resets = []
    aux_bests = []
    for arr in matrix:
        aux_bests = []
        for i in range(len(arr)-1):
            if(arr[i] < arr[i+1]):
                aux_bests.append(arr[i])
            #print(arr[i])
        aux_bests.append(arr[-1])
        best_resets.append(aux_bests)
    return best_resets
```

```python
h1r100 = resets_quality(matrix[0:4])
plt.boxplot(h1r100)
plt.title("Ian10 ; Cant_Hosts = 1 ; Resets = 100")
plt.xlabel("Multiplicador de Penalizaciones")
plt.ylabel("Costo por penalizaciones")
plt.xticks([1, 2, 3, 4], ['CP=1 DP=1 OP=1', 'CP=1 DP=1 OP=5', 'CP=1 DP=5 OP=1', 'CP=5 DP=1 OP=1'], rotation = 30)
plt.show()
```
